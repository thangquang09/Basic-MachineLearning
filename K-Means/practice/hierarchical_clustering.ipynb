# Hierarchical Clustering (Phân cụm theo tầng)

Use the same dataset with kmeans-clustering

## Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Import the dataset

```python
dataset = pd.read_csv('../data/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
dataset.head()
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

The optional number of clusters is the largest distance can you move up and down on the dendrogram plot

```python
img = mpimg.imread('Dendrogram - with line of the optimal number of clusters.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

=> 5 clusters is the best optimal number of clusters, 3 is the second optimal number of clusters

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
print(y_hc)
```

## Visualising the Clusters

```python
colors = ['red', 'blue', 'green', 'violet', 'cyan']

for i in range(0, 5):
    plt.scatter(X[y_hc==i, 0], X[y_hc==i, 1], s = 100, color = colors[i], label = 'Cluster ' + str(i + 1))

plt.title('Clusters of Customers with 5 Clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Try with 3 Clusters

```python
hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
print(y_hc)
```

```python
# Visualising the Clusters
colors = ['red', 'blue', 'green']

for i in range(0, 3):
    plt.scatter(X[y_hc==i, 0], X[y_hc==i, 1], s = 100, color = colors[i], label = 'Cluster ' + str(i + 1))

plt.title('Clusters of Customers with 3 Clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

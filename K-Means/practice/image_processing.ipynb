# Image processing

## Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.cluster import KMeans
```

## Import Image

```python
img = mpimg.imread('../data/tree.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
width = img.shape[0]
height = img.shape[1]
print(img.shape)
```

## Clustering with K-Means

```python
img = img.reshape(width * height, 3)

kmeans = KMeans(n_clusters=4, n_init=10, init='k-means++').fit(img)
labels = kmeans.predict(img)
clusters = kmeans.cluster_centers_
```

```python
print(labels)
print(clusters)
```

```python
img2 = np.zeros_like(img)

for i in range(len(img2)):
    img2[i] = clusters[labels[i]] # Assign per img2 pixel by clusters

img2 = img2.reshape(width, height, 3)
print(img2.shape)
plt.imshow(img2)
plt.axis('off')
plt.show()
```

```python
plt.imsave('Image_Processing.png', img2)
```
